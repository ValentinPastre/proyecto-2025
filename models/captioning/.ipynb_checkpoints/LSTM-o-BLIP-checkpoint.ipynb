#### Codigo para conectar la LSTM con la salida de la inferencia de YOLO
(provisto por chatGPT)

```python
# lstm_modelo.ipynb
import json

# Cargar la salida de YOLO
with open("yolo_output.json", "r") as f:
    yolo_output = json.load(f)

# Usar como entrada de la LSTM
# Por ejemplo, convertir las clases detectadas a secuencia de tokens
tokens = [obj["class"] for obj in yolo_output]
print(tokens)
```
